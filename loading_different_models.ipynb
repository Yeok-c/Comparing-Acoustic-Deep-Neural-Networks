```python
import vggish_keras.vggish_keras.vggish as vgk
from tensorflow.keras import layers, Model

num_classes = 6
vggish = vgk.VGGish(include_top=False)
vggish.load_weights("./vggish_keras/vggish_keras/model/audioset_weights.h5")
# x = layers.GlobalAveragePooling2D()(vggish.layers[-1].output)
x = layers.Dropout(0.4)(vggish.layers[-1].output)
x = layers.Dense(1024)(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(64)(x)
predictions = layers.Dense(num_classes)(x) 
model = Model(vggish.input, predictions)
model.summary()
```

```python
# EfficientNetV1_B0
efficientnet_model = tf.keras.applications.efficientnet.EfficientNetB0(include_top=False, weights=None, input_shape=(96, 64, 1),
pooling=None, classes=num_classes, classifier_activation='softmax')
embeddings = layers.GlobalAveragePooling2D()(efficientnet_model.output)
embeddings = layers.Dropout(0.4)(embeddings)
embeddings = layers.Dense(1024)(embeddings)
embeddings = layers.Dropout(0.2)(embeddings)
embeddings = layers.Dense(64)(embeddings)
logits = layers.Dense(units=num_classes, use_bias=True)(embeddings)
predictions = layers.Activation(activation='Softmax')(logits) # <- Sigmoid
model = Model(efficientnet_model.input, predictions) 
```

```python
# EfficientNetV2_B0
from tensorflow.keras import layers, Model
import tensorflow as tf
num_classes = 6
efficientnetv2_model = tf.keras.applications.efficientnet_v2.EfficientNetV2B0(include_top=False, weights=None, input_shape=(96, 64, 1),
pooling=None, classes=num_classes, classifier_activation='softmax')
# efficientnetv2_model.summary()

# Efficientnet models behave this way. model output can directly be bridged
embeddings = layers.GlobalAveragePooling2D()(efficientnetv2_model.output)
embeddings = layers.Dropout(0.4)(embeddings)
embeddings = layers.Dense(1024)(embeddings)
embeddings = layers.Dropout(0.2)(embeddings)
embeddings = layers.Dense(64)(embeddings)
logits = layers.Dense(units=num_classes, use_bias=True)(embeddings)
predictions = layers.Activation(activation='Softmax')(logits) # <- Sigmoid
model = Model(efficientnetv2_model.input, predictions) 
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
num_classes=6
MobileNetV2 = tf.keras.applications.mobilenet_v2.MobileNetV2(
    input_shape=(96,64,1),
    alpha=1.0,
    include_top=False,
    weights=None,
    input_tensor=None,
    pooling=None,
    classes=6,
    classifier_activation='softmax'
)
# MobileNetV2.summary()
x = layers.GlobalAveragePooling2D()(MobileNetV2.layers[-1].output)
x = layers.Dropout(0.4)(x)
x = layers.Dense(1024)(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(64)(x)
logits = layers.Dense(units=num_classes, use_bias=True)(x)
predictions = layers.Activation(activation='Softmax')(logits) # <- Sigmoid
model = Model(MobileNetV2.input, predictions) 
model.summary()
```

```python
# MobilenetV3
import tensorflow as tf
from tensorflow.keras import layers, Model
num_classes=6
MobileNetV3Large = tf.keras.applications.MobileNetV3Large(
    input_shape=(96,64,1),
    alpha=1.0,
    minimalistic=False,
    include_top=True,
    weights=None,
    input_tensor=None,
    classes=1000,
    pooling=None,
    dropout_rate=0.2,
    classifier_activation='softmax',
    include_preprocessing=True
)
# x = layers.GlobalAveragePooling2D()(MobileNetV1Small.layers[-4].output)
x = layers.Dropout(0.4)(MobileNetV3Large.layers[-4].output)
x = layers.Dense(1024)(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(64)(x)
logits = layers.Dense(units=num_classes, use_bias=True)(x)
predictions = layers.Activation(activation='Softmax')(logits) # <- Sigmoid
model = Model(MobileNetV3Large.input, predictions) 
model.summary()
```
